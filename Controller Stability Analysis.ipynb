$P(A \mid B) = \frac{P(B \mid A)P(A)}{P(B)}$

```python
%%HTML

<div id='wikitext'>
<p class='vspace'>The purpose of controller stability analysis is to determine the range of controller gains between lower `K_{cL}` and upper `K_{cU}` limits that lead to a stable controller.
</p>
<p class='vspace'>$$K_{cL} \le K_c \le K_{cU}$$
</p>
<p class='vspace'>The principles of stability analysis presented here are general for any linear time-invariant system whether it is for controller design or for analysis of system dynamics. Several characteristics of a system in the Laplace domain can be deduced without transforming a system signal or transfer function back into the time domain. Some of the analysis relies on the roots of the transfer function denominator, also known as poles. The roots of the numerator, also known as zeros, do not affect the stability directly but can potentially cancel an unstable pole to create an overall stable system.
</p>
<p class='vspace'><strong>Converge or Diverge</strong>
</p>
<p class='vspace'>A first point of analysis is whether the system converges or diverges. This is determined by analyzing the roots of the denominator of the transfer function. If any of the real parts of the roots of the denominator are positive then the system is unstable.
</p>
<p class='vspace'>A simple rule to determine whether there are positive real roots is to examine the signs of the polynomial. If there are mixed signs (+ or -) then the system will be unstable because there is at least one positive real root. 
</p>
<p class='vspace'>Before modern computational methods, there were several methods devised to determine the stability of a system. One such approach is the <a class='urllink' href='https://en.wikipedia.org/wiki/Routh%e2%80%93Hurwitz_stability_criterion' rel='nofollow'>Routh-Hurwitz stability criterion</a>. The leading left edge of a table determines whether the system is stable for or any <em>n</em>th-degree polynomial
</p>
<p class='vspace'>$$a_n s^n + a_{n-1} s^{n-1} + \cdots + a_1 s + a_0$$
</p>
<p class='vspace'>The coefficients of the polynomial are placed into tabular form and additional coefficients <em>b</em> and <em>c</em> are computed from higher rows. 
</p>
<div class='vspace'></div>
<table border='1' rules='rows' width='100%' align='left' cellpadding='0' ><tr ><td  align='center'>`a_n`</td><td  align='center'>`a_{n-2}`</td><td  align='center'>`a_{n-4}`</td><td  align='center'>`\ldots`</td></tr>
<tr ><td  align='center'>`a_{n-1}`</td><td  align='center'>`a_{n-3}`</td><td  align='center'>`a_{n-5}`</td><td  align='center'>`\ldots`</td></tr>
<tr ><td  align='center'>`b_{1}=\frac{a_{n-1}a_{n-2}-a_n a_{n-3}}{a_{n-1}`</td><td  align='center'>`b_{2}=\frac{a_{n-1}a_{n-4}-a_n a_{n-5}}{a_{n-1}}`</td><td  align='center'>`b_{3}`</td><td  align='center'>`\ldots`</td></tr>
<tr ><td  align='center'>`c_{1}=\frac{b_1 a_{n-3}-a_{n-1} b_{2}}{b_1}`</td><td  align='center'>`c_{2}=\frac{b_1 a_{n-5}-a_{n-1} b_{3}}{b_1}`</td><td  align='center'>`c_{3}`</td><td  align='center'>`\ldots`</td></tr>
<tr ><td  align='center'>`\vdots`</td><td  align='center'>`\vdots`</td><td  align='center'>`\vdots`</td><td  align='center'>`\ddots`</td></tr>
</table>
<p class='vspace'>The terms <em>b</em> and <em>c</em> are:
</p>
<p class='vspace'>$$b_i=\frac{a_{n-1}\times{a_{n-2i}}-a_n\times{a_{n-2i-1}}}{a_{n-1}}$$
</p>
<p class='vspace'>$$c_i=\frac{b_1\times{a_{n-2i-1}}-a_{n-1}\times{b_{i+1}}}{b_1}$$
</p>
<p class='vspace'>A changing sign (+ or -) of the leading left edge `a_n`, `a_{n-1}`, `b_{1}`, `c_{1}` indicates that the system is unstable.
</p>
<p class='vspace'>Several additional methods can be used to determine stability as summarized below.
</p>
<div class='vspace'></div>
<table border='1' rules='rows' width='100%' align='left' cellpadding='0' ><tr ><td  align='left'><strong>Method</strong></td><td  align='left'><strong>Stability Condition</strong></td></tr>
<tr ><td  align='left'>Routh Array</td><td  align='right'>Leading left edge terms are greater than or equal to zero</td></tr>
<tr ><td  align='left'>Root Locus Plot</td><td  align='center'>All poles (roots of denominator) have negative real parts</td></tr>
<tr ><td  align='left'>Bode Plot</td><td  align='center'>The response magnitude at -180 deg phase is less than one</td></tr>
<tr ><td  align='left'>Simulation</td><td  align='center'>Simulate a closed loop response with increasing or decreasing controller gains until instability occurs</td></tr>
</table>
<p class='vspace'>In addition to analysis in the Laplace domain, stability can be determined from a model in <a class='wikilink' href='http://apmonitor.com/pdc/index.php/Main/StateSpaceModel'>state space form</a>.
</p>
<p class='vspace'>$$\dot x = A x + B u$$
</p>
<p class='vspace'>$$y = C x + D u$$
</p>
<p class='vspace'>A state space model is stable when the eigenvalues of the <em>A</em> matrix have negative real parts.
</p>
<p class='vspace'><strong>Oscillatory or Smooth</strong>
</p>
<p class='vspace'>A second point of analysis is whether the system exhibits oscillatory or smooth behavior. If any of the roots of the denominator have an imaginary component then the system has oscillations. Imaginary roots always come in pairs with the same positive and negative imaginary values.
</p>
<p class='vspace'><strong>Final Value Theorem</strong>
</p>
<p class='vspace'>The Final Value Theorem (FVT) gives the steady state gain `K_p` of a transfer function `G(s)` by taking the limit as `s \to 0`
</p>
<p class='vspace'>$$K_p = \lim_{s \to 0}G(s)$$
</p>
<p class='vspace'>The FVT also determines the final signal value `y_\infty` for a stable system with output `Y(s)`. Note that the Laplace variable `s` is multiplied by the signal `Y(s)` before the limit is taken.
</p>
<p class='vspace'>$$y_\infty = \lim_{s \to 0} s \, Y(s)$$
</p>
<p class='vspace'>The FVT may give misleading results if applied to an unstable system. It is only applicable to stable systems or signals.
</p>
<p class='vspace'><strong>Initial Value Theorem</strong>
</p>
<p class='vspace'>The Initial Value Theorem (IVT) gives an initial condition of a signal by taking the limit as `s \to \infty`. Like the FVT, the Laplace variable `s` is multiplied by the signal `Y(s)` before the limit is taken.
</p>
<p class='vspace'>$$y_0 = \lim_{s \to \infty} s \, Y(s)$$
</p>
<p class='vspace'><strong>Controller Stability</strong>
</p>
<p class='vspace'>Controller stability analysis is finding the range of controller gains that lead to a stabilizing controller. There are multiple methods to compute this range between a lower limit `K_{cL}` and an upper limit `K_{cL}`.
</p>
<p class='vspace'>$$K_{cL} \le K_c \le K_{cU}$$
</p>
<p class='vspace'>This range is important to know the range of tuning values that will not lead to a destabilizing controller. With modern computational tools and powerful computers, the simulation based option is frequently used for complex systems.
</p>
<p class='vspace'><strong>Exercise</strong>
</p>
<p class='vspace'>Consider a feedback control system that has the following open loop transfer function.
</p>
<p class='vspace'>$$G(s) = \frac{4K_c}{(s+1)(s+2)(s+3)}$$
</p>
<p class='vspace'>Determine the values of `K_c` that keep the closed loop system response stable.
</p>
<p class='vspace'><strong>Solution</strong>
</p>
<p class='vspace'><em>Routh Array</em>
</p>
<div class='vspace'></div>
<table border='1' rules='rows' width='100%' align='left' cellpadding='0' ><tr ><td  align='center'>`a_n=1`</td><td  align='center'>`a_{n-2}=11`</td></tr>
<tr ><td  align='center'>`$$a_{n-1}=6$$`</td><td  align='center'>`a_{n-3}=6+4 K_c`</td></tr>
<tr ><td  align='center'>`b_{1}=\frac{66 - 6 - 4 K_c}{6}`</td><td  align='center'>`b_{2}=0`</td></tr>
<tr ><td  align='center'>`c_{1}=6+4 K_c`</td><td  align='center'>0</td></tr>
</table>
<p class='vspace'>The leading edge cannot change signs for the system to be stable. Therefore, the following conditions must be met:
</p>
<p class='vspace'>$$a_n=1 &gt; 0$$
</p>
<p class='vspace'>$$a_{n-1}=6 &gt; 0$$
</p>
<p class='vspace'>$$b_{1}=\frac{66 - 6 - 4 K_c}{6} &gt; 0$$
</p>
<p class='vspace'>$$c_{1}=6+4 K_c &gt; 0$$
</p>
<p class='vspace'>The positive constraint on `b_1` leads to `K_c&lt;15`. The positive constraint on `c_1` means that `K_c &gt; -1.5`. Therefore the following ranges are acceptable for the controller stability.
</p>
<p class='vspace'>$$-1.5 &lt; K_c &lt; 15$$
</p>
<p class='vspace'>This is a more comprehensive solution than the other methods shown below because it also includes a lower bound on the controller stability limit (if a direct acting controller were inadvertently used).
</p>
<p class='vspace'><em>Root Locus Plot</em>
</p>
<p class='vspace'>Determine where the real portion of the roots crosses to the right hand side of the plane. In this case, the real part of two roots becomes positive at `K_c=15`.
</p>
<div class='vspace'></div><div class="img"><img width='500px' src='http://apmonitor.com/pdc/uploads/Main/root_locus_plot.png' alt='' title='' /></div>
<p class='vspace'><em>Bode Plot</em>
</p>
<p class='vspace'>Determine the gain margin at -180<sup>o</sup> phase. The magnitude at -180<sup>o</sup> phase is about -23 dB. With `-23 = 20 log_{10} (AR)`, the gain margin is `1/{AR}` and approximately equal to 15. This is the upper bound on the controller gain to keep the system stable. This answer agrees with the root locus plot solution.
</p>
<div class='vspace'></div><div class="img"><img width='500px' src='http://apmonitor.com/pdc/uploads/Main/bode_plot.png' alt='' title='' /></div>
<p class='vspace'><span id='source-tog' class='toggle'><script type="text/javascript">document.write("<input type='button' class='inputbutton togglebutton' value='Show Root Locus and Bode Source Code' onclick=\"javascript:toggleObj('source','show','Show Root Locus and Bode Source Code','Hide','Show','Hide','','','','1','', 'block')\" />")</script></span>
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
# open loop
num = [4.0]
den = [1.0,6.0,11.0,6.0]
sys = signal.TransferFunction(num, den)
t1,y1 = signal.step(sys)

# closed loop
Kc = 1.0
num = [4.0*Kc]
den = [1.0,6.0,11.0,4.0*Kc+6.0]
sys2 = signal.TransferFunction(num, den)
t2,y2 = signal.step(sys2)

plt.figure(1)
plt.subplot(2,1,1)
plt.plot(t1,y1,'k-')
plt.legend(['Open Loop'],loc='best')

plt.subplot(2,1,2)
plt.plot(t2,y2,'r--')
plt.legend(['Closed Loop'],loc='best')
plt.xlabel('Time')

# root locus plot
import numpy.polynomial.polynomial as poly
n = 1000  # number of points to plot
nr = 3    # number of roots
rs = np.zeros((n,2*nr))   # store results
Kc = np.logspace(-2,2,n)  # Kc values
for i in range(n):        # cycle through n times
    den = [1.0,6.0,11.0,4.0*Kc[i]+6.0] # polynomial
    roots = poly.polyroots(den) # find roots
    for j in range(nr):   # store roots
        rs[i,j] = roots[j].real # store real
        rs[i,j+nr] = roots[j].imag # store imaginary
plt.figure(2)
plt.subplot(2,1,1)
plt.xlabel('Root (real)')
plt.ylabel('Root (imag)')
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--')
for i in range(nr):
    plt.plot(rs[:,i],rs[:,i+nr],'.')
plt.subplot(2,1,2)
plt.plot([Kc[0],Kc[-1]],[0,0],'k-')
for i in range(3):
    plt.plot(Kc,rs[:,i],'.')
plt.ylabel('Root (real part)')
plt.xlabel('Controller Gain (Kc)')

# bode plot
w,mag,phase = signal.bode(sys)
plt.figure(3)
plt.subplot(2,1,1)
plt.semilogx(w,mag,'k-',linewidth=3)
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--')
plt.ylabel('Magnitude')

plt.subplot(2,1,2)
plt.semilogx(w,phase,'k-',linewidth=3)
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--')
plt.ylabel('Phase')
plt.xlabel('Frequency')

plt.show()
```
