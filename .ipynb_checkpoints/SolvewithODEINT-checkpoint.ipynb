# Solve Differential Equations with ODEINT
<div id='wikitext'>
<p class='vspace'>Differential equations are solved in Python with the Scipy.integrate package using function <strong>ODEINT</strong>. Another Python package that solves differential equations is <a class='urllink' href='https://gekko.readthedocs.io/en/latest/' rel='nofollow'>GEKKO</a>. See this <a class='wikilink' href='http://apmonitor.com/pdc/index.php/Main/PythonDifferentialEquations'>link for the same tutorial in GEKKO</a> versus ODEINT. ODEINT requires three inputs:
</p>
<div class='vspace'></div><pre> y = odeint(model, y0, t)
</pre><div class='vspace'></div><ol><li><strong>model</strong>: Function name that returns derivative values at requested y and t values as dydt = model(y,t)
</li><li><strong>y0</strong>: Initial conditions of the differential states
</li><li><strong>t</strong>: Time points at which the solution should be reported. Additional internal points are often calculated to maintain accuracy of the solution but are not reported.
</li></ol><div class='vspace'></div>


<p class='vspace'>An example of using <em>ODEINT</em> is with the following differential equation with parameter <em>k=0.3</em>, the initial condition <em>y<sub>0</sub>=5</em> and the following differential equation.
</p>
<p class='vspace'>$$\frac{dy(t)}{dt} = -k \; y(t)$$
</p>
<p class='vspace'>The Python code first imports the needed Numpy, Scipy, and Matplotlib packages. The model, initial conditions, and time points are defined as inputs to <em>ODEINT</em> to numerically calculate <em>y(t)</em>.
</p>

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

# function that returns dy/dt
def model(y,t):
    k = 0.3
    dydt = -k * y
    return dydt

# initial condition
y0 = 5

# time points
t = np.linspace(0,20)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

An optional fourth input is args that allows additional information to be passed into the model function. The args input is a tuple sequence of values. The argument k is now an input to the model function by including an addition argument.

```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t,k):
    dydt = -k * y
    return dydt

# initial condition
y0 = 5

# time points
t = np.linspace(0,20)

# solve ODEs
k = 0.1
y1 = odeint(model,y0,t,args=(k,))
k = 0.2
y2 = odeint(model,y0,t,args=(k,))
k = 0.5
y3 = odeint(model,y0,t,args=(k,))

# plot results
plt.plot(t,y1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,y3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

<p class='vspace'><strong>Problem 1</strong>
</p>
<p class='vspace'>$$\frac{dy(t)}{dt} = -y(t) + 1$$
</p>
<p class='vspace'>$$y(0) = 0$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t):
    dydt = -y + 1.0
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0,5)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

<p class='vspace'><strong>Problem 2</strong>
</p>
<p class='vspace'>$$5 \; \frac{dy(t)}{dt} = -y(t) + u(t)$$
</p>
<p class='vspace'>$$y(0) = 1$$
</p>
<p class='vspace'  style='text-align: center;'>`u` steps from <em>0</em> to <em>2</em> at `t=10` 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(y,t):
    # u steps from 0 to 2 at t=10
    if t<10.0:
        u = 0
    else:
        u = 2
    dydt = (-y + u)/5.0
    return dydt

# initial condition
y0 = 1

# time points
t = np.linspace(0,40,1000)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y,'r-',label='Output (y(t))')
plt.plot([0,10,10,40],[0,0,2,2],'b-',label='Input (u(t))')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```

<p class='vspace'><strong>Problem 3</strong>
</p>
<p class='vspace'>Solve for `x(t)` and `y(t)` and show that the solutions are equivalent.
</p>
<p class='vspace'>$$\frac{dx(t)}{dt} = 3 \; exp(-t)$$
$$\frac{dy(t)}{dt} = 3 - y(t)$$
$$x(0) = 0$$
$$y(0) = 0$$
</p>

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def model(z,t):
    dxdt = 3.0 * np.exp(-t)
    dydt = -z[1] + 3
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0,0]

# time points
t = np.linspace(0,5)

# solve ODE
z = odeint(model,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label=r'$\frac{dx}{dt}=3 \; \exp(-t)$')
plt.plot(t,z[:,1],'r--',label=r'$\frac{dy}{dt}=-y+3$')
plt.ylabel('response')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```

<p class='vspace'><strong>Problem 4</strong>
</p>
<p class='vspace'>$$2 \; \frac{dx(t)}{dt} = -x(t) + u(t)$$
</p>
<p class='vspace'>$$5 \; \frac{dy(t)}{dt} = -y(t) + x(t)$$
</p>
<p class='vspace'>$$u = 2 \, S(t-5), \; x(0) = 0, \; y(0) = 0$$
</p>
<p class='vspace'>where `S(t-5)` is a step function that changes from zero to one at `t=5`. When it is multiplied by two, it changes from zero to two at that same time, `t=5`.
</p>

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def model(z,t,u):
    x = z[0]
    y = z[1]
    dxdt = (-x + u)/2.0
    dydt = (-y + x)/5.0
    dzdt = [dxdt,dydt]
    return dzdt

# initial condition
z0 = [0,0]

# number of time points
n = 401

# time points
t = np.linspace(0,40,n)

# step input
u = np.zeros(n)
# change to 2.0 at time = 5.0
u[51:] = 2.0

# store solution
x = np.empty_like(t)
y = np.empty_like(t)
# record initial conditions
x[0] = z0[0]
y[0] = z0[1]

# solve ODE
for i in range(1,n):
    # span for next time step
    tspan = [t[i-1],t[i]]
    # solve for next step
    z = odeint(model,z0,tspan,args=(u[i],))
    # store solution for plotting
    x[i] = z[1][0]
    y[i] = z[1][1]
    # next initial condition
    z0 = z[1] 

# plot results
plt.plot(t,u,'g:',label='u(t)')
plt.plot(t,x,'b-',label='x(t)')
plt.plot(t,y,'r--',label='y(t)')
plt.ylabel('values')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```
